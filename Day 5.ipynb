### Linked List

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def add_from_back(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        cur = self.head
        while cur.next:
            cur = cur.next
        cur.next = new_node

    def add_from_front(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def delete_from_back(self):
        if not self.head:
            print("List is empty")
            return
        if not self.head.next:
            self.head = None
            return
        cur = self.head
        while cur.next.next:
            cur = cur.next
        cur.next = None

    def delete_from_front(self):
        if not self.head:
            print("List is empty")
            return
        self.head = self.head.next

    def display(self):
        cur = self.head
        while cur:
            print(cur.data, end=" -> ")
            cur = cur.next
        print("None")

    def search_element(self, data):
        cur = self.head
        position = 0
        while cur:
            if cur.data == data:
                return position
            cur = cur.next
            position += 1
        return -1
    def update_list(self, n, data):
        cur=self.head
        i=0
        while i <=n and cur!=None:
            i+=1
            cur=cur.next
        cur.data=data
        return
    def rotate_list(self):
        cur=self.head
        prev=None
        while cur.next is not None:
            prev=cur
            cur=cur.next
        prev.next=None
        cur.next=self.head
        self.head=cur
        
    def print_reverse(self):
        def _print_reverse(node):
            if node is None:
                return
            _print_reverse(node.next)
            print(node.data, end=" -> ")
        _print_reverse(self.head)
        print("None")
        
#print reverse in linked list using recursion
#remove all occurences of linked list another linked list
```

```python
def main():
    ll = LinkedList()
    while True:
        print("Choose an operation:")
        print("1: Add from back")
        print("2: Add from front")
        print("3: Delete from back")
        print("4: Delete from front")
        print("5: Display")
        print("6: Search element")
        print("7: Update Node")
        print("8: Move last node to first")
        print("9: Reverse print ")
        print("10: Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            data = int(input("Enter data to add from back: "))
            ll.add_from_back(data)
            
        elif choice == '2':
            data = int(input("Enter data to add from front: "))
            ll.add_from_front(data)
            
        elif choice == '3':
            ll.delete_from_back()
            
        elif choice == '4':
            ll.delete_from_front()
            
        elif choice == '5':
            ll.display()
            
        elif choice == '6':
            data = int(input("Enter data to search: "))
            pos = ll.search_element(data)
            if pos != -1:
                print(f"Element {data} found at position {pos}")
            else:
                print(f"Element {data} not found")
        elif choice == '7':
            n=int(input("Enter position to be updated: "))
            data=int(input("Enter the new data: "))
            ll.update_list(n,data)
        elif choice=='8':
            ll.rotate_list()
        elif choice=='9':
            ll.print_reverse()
        elif choice=='10':
            break
        else:
            print("Invalid choice, please try again.")

if __name__ == "__main__":
    main()
```

```python
"""
Remove All Occurrences of a Sublist in Another Linked List
"""
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def remove_all_occurrences(head, sub_head):
    if not head or not sub_head:
        return head

    # Helper function to check if two lists are equal starting from given nodes
    def is_equal(start1, start2):
        while start1 and start2:
            if start1.data != start2.data:
                return False
            start1 = start1.next
            start2 = start2.next
        return start2 is None

    dummy = Node(0)
    dummy.next = head
    prev = dummy
    cur = head

    while cur:
        if is_equal(cur, sub_head):
            # Remove the sublist
            sublist_cur = sub_head
            while sublist_cur:
                cur = cur.next
                sublist_cur = sublist_cur.next
            prev.next = cur
        else:
            prev = cur
            cur = cur.next

    return dummy.next

# Example usage:
main_list = LinkedList()
main_list.add_from_back(1)
main_list.add_from_back(2)
main_list.add_from_back(3)
main_list.add_from_back(2)
main_list.add_from_back(4)
main_list.display()

sub_list = LinkedList()
sub_list.add_from_back(2)

new_head = remove_all_occurrences(main_list.head, sub_list.head)
main_list.head = new_head
main_list.display()  # Should print 1 -> 3 -> 4 -> None
```
